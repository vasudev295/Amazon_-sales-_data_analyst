```python
top10state = df['ship-state'].value_counts().head(10) 

plt.figure(figsize=(12,6)) 
sns.countplot(data=df[df['ship-state'].isin(top10state)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of top states')
plt.xticks(rotation=45) 
plt.show() 
```

```python

top10state = df['ship-state'].value_counts().head(10)

# Filter the DataFrame to include only the top 10 states
df_top10 = df[df['ship-state'].isin(top10state.index)]

# Plotting
plt.figure(figsize=(12, 6))
sns.countplot(data=df_top10, x='ship-state', order=top10state.index)
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.title('Distribution of top states')
plt.xticks(rotation=45) 
plt.show()  
```
